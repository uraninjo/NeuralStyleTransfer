```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model=hub.load("https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2")
```

```python
def img_yükleme(path):
    img=tf.io.read_file(path)
    img=tf.image.decode_image(img,channels=3)
    img=tf.image.convert_image_dtype(img,tf.float32)
    img=img[tf.newaxis,:]
    return img
```

```python
dönüştürülecek=img_yükleme('photos\\dönüştürülecek\\11.jpg')
stil=img_yükleme('photos\\stil\\vegeta.jpg')
```

```python
plt.imshow(np.squeeze(dönüştürülecek))
plt.show()
```

```python
plt.imshow(np.squeeze(stil))
plt.show()
```

```python
dönüştürülmüş_img=model(tf.constant(dönüştürülecek),tf.constant(stil))[0]
```

```python
plt.imshow(np.squeeze(dönüştürülmüş_img))
plt.show()
```

```python
cv2.imwrite('generated\\generated.jpg',cv2.cvtColor(np.squeeze(dönüştürülmüş_img)*255,cv2.COLOR_BGR2RGB))
```
